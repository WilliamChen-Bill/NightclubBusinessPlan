## 2. Mathematical model (a) ##

* **part a**
* In this economic question. We will need the concept of the multiperiod planning programming to calculate the minimun cost to product the different drinks in our club.

* $x_{i,t}$ represent for the $i$ drinks production in $t$ month
* $b_t$ represent for number of bartenders be employed in $t$month
* $I_{i,t}$ represent for the inventory of the $i$ drinks not usde in $t$ month
* $h_{t}$ represent for number of new hired bartender in $t$ month
* $f_{t}$ represent for number of fired bartender in $month$
* $b_0 = 10$
* $d_{i,t}$ represent for the $i$ drinks demand in $t$ month

* math model:
$$
\underset{t}{min}\:\: \sum (C_ix_{i,t}+ C_{\mbox{bartender}}(b_t+ h_t)+C_{\mbox{fired bartender}}f_t-C_iI_t)\\
s.t. \:\:\mbox{time cost}_ix_i\le b_t*64*60\\
I_{t-1}+x_t=d_{i,t}+I_t\\
b_t=b_{t-1}+h_t-f_t\\
x_{i,t},d_{i,t},b_t,h_t,f_t\ge 0\\
i \in \{\mbox{different drinks}\}\\
t \in \{1,2,3,4,5,6\}
$$
* The reason inventory can is negative is bc alcohol can be store won't rotton (we assume), so we can sell it in next month.

```julia
# part a
using JuMP, Gurobi
m =Model(Gurobi.Optimizer)
set_optimizer_attribute(m, "OutputFlag", 0)

amount = (length(demand1))/8
D1 = (demand1' * ones(Int, Int(amount)))' # total demand1 array
amount = (length(demand2))/8
D2 = (demand2' * ones(Int, Int(amount)))' # total demand2 array
amount = (length(demand3))/8
D3 = (demand3' * ones(Int, Int(amount)))' # total demand3 array
amount = (length(demand4))/8
D4 = (demand4' * ones(Int, Int(amount)))' # total demand4 array
amount = (length(demand5))/8
D5 = (demand5' * ones(Int, Int(amount)))' # total demand5 array
amount = (length(demand6))/8
D6 = (demand6' * ones(Int, Int(amount)))' # total demand6 array
D = vcat(D1,D2,D3,D4,D5,D6)


@variable(m, x1[1:6] >= 0, Int) # drinks1 produced in t=1,2,3,4,5,6
@variable(m, x2[1:6] >= 0, Int) # drinks2 produced in t=1,2,3,4,5,6
@variable(m, x3[1:6] >= 0, Int) # drinks3 produced in t=1,2,3,4,5,6
@variable(m, x4[1:6] >= 0, Int) # drinks4 produced in t=1,2,3,4,5,6
@variable(m, x5[1:6] >= 0, Int) # drinks5 produced in t=1,2,3,4,5,6
@variable(m, x6[1:6] >= 0, Int) # drinks6 produced in t=1,2,3,4,5,6
@variable(m, x7[1:6] >= 0, Int) # drinks7 produced in t=1,2,3,4,5,6
@variable(m, x8[1:6] >= 0, Int) # drinks8 produced in t=1,2,3,4,5,6


@variable(m, b[1:6] >= 0, Int) # bartenders employed
@variable(m, h[1:6] >= 0, Int) # bartenders hired 
@variable(m, f[1:6] >= 0, Int) # bartenders fired

@variable(m, I1[1:6]>=0, Int) # inventory for different drinks in 1-6 month
@variable(m, I2[1:6]>=0, Int)
@variable(m, I3[1:6]>=0, Int)
@variable(m, I4[1:6]>=0, Int)
@variable(m, I5[1:6]>=0, Int)
@variable(m, I6[1:6]>=0, Int)
@variable(m, I7[1:6]>=0, Int)
@variable(m, I8[1:6]>=0, Int)


@expression(m, drinks_cost,sum(x1) * drinks_cost_price[1] 
    + sum(x2)*drinks_cost_price[2]
    + sum(x3)*drinks_cost_price[3]
    + sum(x4)*drinks_cost_price[4]
    + sum(x5)*drinks_cost_price[5]
    + sum(x6)*drinks_cost_price[6]
    + sum(x7)*drinks_cost_price[7]
    + sum(x8)*drinks_cost_price[8])

@expression(m, inv_save, sum(I1) * drinks_cost_price[1] 
    + sum(I2)*drinks_cost_price[2]
    + sum(I3)*drinks_cost_price[3]
    + sum(I4)*drinks_cost_price[4]
    + sum(I5)*drinks_cost_price[5]
    + sum(I6)*drinks_cost_price[6]
    + sum(I7)*drinks_cost_price[7]
    + sum(I8)*drinks_cost_price[8])

@objective(m, Min, drinks_cost + bartender_wage*(sum(b)+sum(h) + 200*sum(f))-inv_save)

@constraint(m, produce[t in 1:6], time_produce[1]*x1[t]
    +time_produce[2]*x2[t]
    +time_produce[3]*x3[t]
    +time_produce[4]*x4[t]
    +time_produce[5]*x5[t]
    +time_produce[6]*x6[t]
    +time_produce[7]*x7[t]
    +time_produce[8]*x8[t] <= 64*60*b[t])
@constraint(m, x1[1] == D[1,1]+I1[1])
@constraint(m, x2[1] == D[1,2]+I2[1])
@constraint(m, x3[1] == D[1,3]+I3[1])
@constraint(m, x4[1] == D[1,4]+I4[1])
@constraint(m, x5[1] == D[1,5]+I5[1])
@constraint(m, x6[1] == D[1,6]+I6[1])
@constraint(m, x7[1] == D[1,7]+I7[1])
@constraint(m, x8[1] == D[1,8]+I8[1])

@constraint(m, Inv_bal_d1[t in 2:6],I1[t-1]+x1[t] == D[t,1]+I1[t])
@constraint(m, Inv_bal_d2[t in 2:6],I2[t-1]+x2[t] == D[t,2]+I2[t])
@constraint(m, Inv_bal_d3[t in 2:6],I3[t-1]+x3[t] == D[t,3]+I3[t])
@constraint(m, Inv_bal_d4[t in 2:6],I4[t-1]+x4[t] == D[t,4]+I4[t])
@constraint(m, Inv_bal_d5[t in 2:6],I5[t-1]+x5[t] == D[t,5]+I5[t])
@constraint(m, Inv_bal_d6[t in 2:6],I6[t-1]+x6[t] == D[t,6]+I6[t])
@constraint(m, Inv_bal_d7[t in 2:6],I7[t-1]+x7[t] == D[t,7]+I7[t])
@constraint(m, Inv_bal_d8[t in 2:6],I8[t-1]+x8[t] == D[t,8]+I8[t])

@constraint(m, work_bal1, 10 - f[1]+h[1]==b[1])
@constraint(m, work_bal[t in 2:6], b[t-1]-f[t]+h[t]==b[t])
```

## 3. Solution (a) ##

```julia
optimize!(m)
minimum_cost = objective_value(m)
production_drink1 = value.(x1')
production_drink2 = value.(x2')
production_drink3 = value.(x3')
production_drink4 = value.(x4')
production_drink5 = value.(x5')
production_drink6 = value.(x6')
production_drink7 = value.(x7')
production_drink8 = value.(x8')

println("minimum cost: \$", objective_value(m))
println("bartender employed: ", value.(b'))
println("Martini produce: ", value.(x1'))
println("Cosmopolitan produce: ", value.(x2'))
println("Pink Lady produce: ", value.(x3'))
println("Sex on the Beach produce: ", value.(x4'))
println("Sea Breeze produce: ", value.(x5'))
println("All Shot produce: ", value.(x6'))
println("Bears produce: ", value.(x7'))
println("Fountain Drinks produce: ", value.(x8'))
println("inventory Martini: ", value.(I1'))
println("inventory Cosmopolitan: ", value.(I2'))
println("inventory Pink Lady: ", value.(I3'))
println("inventory Sex on the Beach: ", value.(I4'))
println("inventory Sea Breeze: ", value.(I5'))
println("inventory All Shot produce: ", value.(I6'))
println("inventory Bears produce: ", value.(I7'))
println("inventory Fountain Drinks produce: ", value.(I8'))
```

## 4. Mathematical model (b) ##
* **part b**
* in this part we will need to use logical constraint to make a decision about to open nightclub or bar in each month. This will utilize the integer programming to make a decision.

* math model.
* $t$ = month
* $P_{b,t}$ = bar profit 
* $C_{b,t}$ = total cost for bar each month
* $z_t$ = decision for bar to open in each month
* $P_{n,t}$ = nightclub profit
* $C_{n,t}$ = total_cost_nightclub_month
* $\mu_t$ = decision for nightclub to open in each month
* $P_{\mbox{drinks}, t}$ = drinks profit in each month
$$
max \sum(P_{b,t}-C_{b,t}z_t)+\sum(P_{n,t}-C_{n,t}\mu_t)\\
P_{b,t},C_{b,t},P_{n,t},C_{n,t}\ge 0\\
\mu_t \in \{0,1\}\\
z_t \in \{0,1\}\\
z_t+\mu_t=1\\
P_{b,t}\le \frac{P_{\mbox{drinks}, t}}{5}*\mu_t\\
P_{n,t}\le P_{\mbox{drinks}, t}*z_t\\
t \in \{1,2,3,4,5,6\}\\
$$


## 5. Solution (b) ##

```julia
using JuMP, Gurobi
m = Model(Gurobi.Optimizer)
set_optimizer_attribute(m, "OutputFlag", 0)

DJ_equipment_cost = sum(DJ_equipment)
bar_equipment_cost = sum(bar_equipment_cost)
bar_monthly_cost = 6*sum(bar_monthly_cost)
nightclub_equipment_cost = sum(nightclub_equipment_cost)
nightclub_monthly_cost = 6*sum(nightclub_monthly_cost)
bar_space_rent = 23.93*500
nightclub_space_rent = 23.93*2500
bar_maintainace = 200
nightclub_maintainace = 1000
admission_profit = 100*total_student

total_cost_bar_month =(minimum_cost/6)/5+bar_maintainace/6+bar_monthly_cost/6+bar_space_rent/6+bar_equipment_cost/6
total_cost_nightclub_month = minimum_cost/6+nightclub_equipment_cost/6+nightclub_monthly_cost/6+nightclub_space_rent/6+nightclub_maintainace/6

@variable(m, bar_profit[1:6] >= 0)
@variable(m, nightclub_profit[1:6] >= 0)
@variable(m, z[1:6] >= 0, Bin)
@variable(m, u[1:6] >= 0, Bin)
@objective(m, Max,sum(bar_profit[i]-total_cost_bar_month*z[i] for i in 1:6)+sum(nightclub_profit[i]-total_cost_nightclub_month*u[i] for i in 1:6))

@expression(m, drinks_profit[i in 1:6], production_drink1[i] * drinks_sell_price[1] 
    + production_drink2[i]*drinks_sell_price[2]
    + production_drink3[i]*drinks_sell_price[3]
    + production_drink4[i]*drinks_sell_price[4]
    + production_drink5[i]*drinks_sell_price[5]
    + production_drink6[i]*drinks_sell_price[6]
    + production_drink7[i]*drinks_sell_price[7]
    + production_drink8[i]*drinks_sell_price[8])

@constraint(m, bar_con[i in 1:6],bar_profit[i] <= drinks_profit[i]/5*z[i])
@constraint(m, nightclub_con[i in 1:6],nightclub_profit[i] <= drinks_profit[i]*u[i])
@constraint(m, logic_con[i in 1:6],sum(z[i]+u[i]) <= 1)

optimize!(m)
println(objective_value(m))
println("decision to open bar for each month", value.(z))
println("decision to open nightclub for each month", value.(u))
println("revenue to open bar for each month",value.(bar_profit))
println("revenue to open nightclub for each month",value.(nightclub_profit))
```
